```python
import re 

def compte_mot(mot):
    L = [] #liste with the number of swearwords per year
    total = 0
    for k in range (0,21): #we go through the years from 2000 to 2020
        S=0 #the sum of words  for the year 200k
        for i in range (0,10) : #we go through the ten songs of the years from 2000 to 2020
            lyrics=list5
            result = len(re.findall(mot,lyrics[10*k+i]))
            #print(lyrics[10*k+i]) 
            #print(result)
            S=S+result  # we add the number of words from song i to th ethe total of the words for the year 200k 
            total=total+result # total for all the years
        L.append(S)
    return(mot,L,total) 

list_words=['fuck','shit','cock','cunt', 'ass','asshole', 'motherfuck','cocksucker','penis','vagina', 'tit', 'breast', 'boob', 'butt', 'pussy', 'pussies', 'tush','bitch','nigg', 'faggot' ]

decompte=[]
for k in range(0,len(list_words)):
    decompte.append(compte_mot(list_words[k])) 
print(decompte)
```
