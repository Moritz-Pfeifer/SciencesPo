On June 24th 2021 the **UEFA** (Union of European Football Associations), announced [its decision to abolish the away goals rule in all UEFA club competitions](https://www.uefa.com/returntoplay/news/026a-1298aeb73a7a-5b64cb68d920-1000--abolition-of-away-goals-rule-in-all-uefa-club-competitions/) as of the 21/22 season. 
This rule was applied to determine the winner of a two-legged knockout tie in cases where the two teams had scored the same number of goals on aggregate over the two matches. In such cases, the team which had scored the higher number of goals away from home was considered the winner of the tie and qualified for the next round of the competition. 

![https://upload.wikimedia.org/wikipedia/fr/b/b5/UEFA_logo.svg](https://upload.wikimedia.org/wikipedia/fr/b/b5/UEFA_logo.svg)
    
It was <u>first applied</u> in the UEFA Cup Winners' Cup when Budapest Honvéd beat Dukla Prague in the second round in 1965–66, and has accompanied us personally throughout our childhood and youth and either helped our favourite teams to qualify for the next round or led to their elimination from the competition. It was founded on the presumption that <u>playing at home would be beneficial</u> in comparison to playing away. The inconvenience of travelling for hours, of having to stay several nights abroad, of not being used to the field’s characteristics, as well as the presence of the opposite spectators played hardly against the away teams. That is why the proportion of home/away wins was of 61%/19%, which, according to the UEFA has now been reduced to 47%/30% (it does not add up to 100% because we also have to take into account the draws)
In order to justify its decision and to provide an explanation for the gap’s reduction, the UEFA made the following statement: 
    
_'Many different factors may be considered as having an impact on this decline in home advantage. Better pitch quality and standardised pitch sizes, improved stadium infrastructure, higher security conditions, enhanced care of refereeing (and more recently the introduction of technological support such as GLT and VAR), wider and more sophisticated TV coverage of matches, more comfortable travel conditions, a compressed calendar dictating squad turnover, and changes in competition formats are all elements which have affected the way football is played and blurred the lines between playing at home and away.'_
    
However, there is a <u>factor that has not changed at all</u> during the past 60 years: **supporters**. Although pitch and travel conditions may have improven, away teams have always had to face the thunderous roar of the home supporters who overwhelmingly exceed the number of away fans who are able to travel to support their team and are allowed into the stadiums.

    
![https://cdn.techexplorist.com/wp-content/uploads/2020/01/football-fan-1392x667.jpg](https://cdn.techexplorist.com/wp-content/uploads/2020/01/football-fan-1392x667.jpg)
    
    
As all other factors can now be considered as equal between home or away, we wondered what impact supporters would have on the results of their teams. 

Thanks to the _COVID-19-crisis_, a door opened for us. With soccer competitions being held behind closed doors during 6 to 12 months between 2020 and 2021, we disposed of a large sample of matches without spectators to compare with the ones in which home teams had been able to feel the warmth and the support of their fans.
 
Thus,


**<center><u>to what extent does the presence of the home-supporters conditionate the winning probabilities of a soccer team?</center></u>**

```python
# stack all the already stacked Excel files of the Top 5 First Leagues in order to create an aggregated database

import os 
import pandas as pd
import glob

all_data = []
for f in glob.glob("/Users/javi/Desktop/Python/FIRST_LEAGUES/FINALS/*.xlsx"):
    all_data.append(pd.read_excel(f))
    
dfraw = pd.concat(all_data, ignore_index=True)

writer = pd.ExcelWriter('final.xlsx')
dfraw.to_excel(r"/Users/javi/Desktop/Python/FIRST_LEAGUES/AGGREGATE_FIRST_LEAGUES.xlsx", index = False)
writer.save()
```

```python
dfraw
```

```python
# We will rename the columns to make them more understandable

names = {"Date": "Date",
         "HomeTeam": "HomeTeam",
         "AwayTeam": "AwayTeam",
         "FTHG": "HomeGoals",
         "FTAG": "AwayGoals",
         "FTR": "Result",
         "Unnamed: 8": "Spectators"
        }

dfraw.rename(columns=names, inplace=True)
dfraw
```

```python
# Now we will replace NaN with yes, to show that they were spectators (in contrast to no where there werent)

dfraw["Spectators"].fillna('yes', inplace=True)
dfraw
```

```python
# Now we create an additional column which shows the goal difference (another way of showing who has won or lost)
difference = dfraw["HomeGoals"] - dfraw["AwayGoals"]
dfraw['GoalDifference'] = difference
dfraw
```

```python
# Now we create a new cleaned data frame which only shows following columns, the ones that are interesting for us.
# In reality, we would only need the 'Result' and the 'Spectators' columns, but we keep the others as well
# We save it into our computer

dfclean = dfraw[['Date', 'HomeTeam', 'AwayTeam', 'HomeGoals', 'AwayGoals', 'Result', 'Spectators', 'GoalDifference']]

dfclean.to_excel(r"/Users/javi/Desktop/Python/FIRST_LEAGUES/AGGREGATE_FIRST_LEAGUES.xlsx", index = False)
dfclean
```

```python
# Right now we eliminate all the draws, as they will not help us further 
dfnodraws = dfclean[dfclean.GoalDifference != 0]
dfnodraws
```

```python
# now we sort our table by the dates of the matches
dfnodraws.sort_values(by="Date")
```

```python
# Now we create a new file only with the matches without spectators
dfNO = dfnodraws[~dfnodraws.Spectators.str.contains("yes")]

# and a new file only with the matches with spectators
dfYES = dfnodraws[~dfnodraws.Spectators.str.contains("no")]
```

```python
# Percentage calculation with spectators
percentageyes = (dfYES['Result'].value_counts()/dfYES['Result'].count())*100
percentageyes
```

```python
# Percentage calculation without spectators
percentageno = (dfNO['Result'].value_counts()/dfNO['Result'].count())*100
percentageno
```

```python
# Hereafter we will display two charts to make the comparison visually more attractive

xYES = (dfYES['Result'].value_counts()/dfYES['Result'].count())*100
label = ["Home", "Away"]
xNO = (dfNO['Result'].value_counts()/dfNO['Result'].count())*100
label = ["Home", "Away"]

plt.subplot(1, 2, 1)
plt.pie(xYES, labels=label)
plt.title('With spectators')

plt.subplot(1, 2, 2)
plt.pie(xNO, labels=label)
plt.title('Without spectators')
plt.tight_layout()
plt.show()

# we will now center the displayed charts
from IPython.core.display import HTML
HTML("""
<style>
.output_png {
    display: table-cell;
    text-align: center;
    vertical-align: middle;
}
</style>
""")

```

```python
means = [dfYES['HomeGoals'].mean(), dfNO['HomeGoals'].mean(), dfYES['AwayGoals'].mean(),dfNO['AwayGoals'].mean()]
categories = ['Home goals WITH spectators', 'Home goals WITHOUT spectators', 'Away goals WITH spectators', 'Away goals WITHOUT spectators']

averagegoals = pd.Series(data=means, index=categories, name="Average goals")
averagegoals
```

```python
# data to plot
n_groups = 2
means_yes = (dfYES['HomeGoals'].mean(), dfYES['AwayGoals'].mean())
means_no = (dfNO['HomeGoals'].mean(), dfNO['AwayGoals'].mean())

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.7

rects1 = plt.bar(index, means_yes, bar_width,
alpha=opacity,
color='b',
label='Spectators')

rects2 = plt.bar(index + bar_width, means_no, bar_width,
alpha=opacity,
color='r',
label='No spectators')

plt.xlabel('Goals')
plt.ylabel('Average')
plt.title('Goal comparison')
plt.xticks(index + 1/2*bar_width, ('Home Goals', 'Away Goals'))
plt.legend()

plt.tight_layout()
plt.show()
```
