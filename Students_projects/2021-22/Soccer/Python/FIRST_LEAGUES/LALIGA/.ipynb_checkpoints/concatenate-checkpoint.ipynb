```python
import os 
import pandas as pd
import glob

all_data = []
for f in glob.glob("/Users/javi/Desktop/Python/LALIGA/*.xlsx"):
    all_data.append(pd.read_excel(f))
    
df = pd.concat(all_data, ignore_index=True)

writer = pd.ExcelWriter('final.xlsx')
df.to_excel(r"/Users/MoritzPfeifer/Documents/LALIGA/Libro.xlsx", index = False)
writer.save()
```
