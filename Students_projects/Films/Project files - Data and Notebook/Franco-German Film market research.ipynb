# GERMAN MOVIES, MINORITY CO-PRODUCTIONS *INCLUDED*



```python
#   We read into our dataframe the overview for German movies, minority coproductions included
GermanMoviesMinorityIncluded = pd.DataFrame([])
for i in range(0, 4): 
    path = working_directory + "/DatengrundlageEUOBS/Deutsche_Filme Daten_Minority_Co-Productions_included/Overview_german_films_2000-2021_include_minority_producing_country/lumiere-search-search(" + str(i) + ").csv"
    locals()["dflocal" + str(i)]= pd.read_csv(path, sep=";", engine="python", nrows = 50)
    GermanMoviesMinorityIncluded = GermanMoviesMinorityIncluded.append(locals()["dflocal" + str(i)], ignore_index=True)
```

```python
#   We clean the the total admissions column in the overview field to later add it. 
GermanMoviesMinorityIncluded["Admissions EUR EU (since 1996)"] = (GermanMoviesMinorityIncluded["Admissions EUR EU (since 1996)"]).fillna(0)
GermanMoviesMinorityIncluded["Admissions EUR EU (since 1996)"]= (GermanMoviesMinorityIncluded["Admissions EUR EU (since 1996)"]).str.replace(' ', '').astype("float64")
GermanMoviesMinorityIncluded["Admissions EUR EU (since 1996)"] = (GermanMoviesMinorityIncluded["Admissions EUR EU (since 1996)"]).fillna(0)
GermanMoviesMinorityIncluded["Admissions EUR EU (since 1996)"] = (GermanMoviesMinorityIncluded["Admissions EUR EU (since 1996)"]).round(0).astype("int")
```

```python
#  We sum the total of German movies, minority included
SUMGermanMoviesMinorityIncluded = sum(GermanMoviesMinorityIncluded["Admissions EUR EU (since 1996)"])
```

```python
#   
print("The total admissions of top 200 German movies in the EU, minority coproductions included is:", sum(GermanMoviesMinorityIncluded["Admissions EUR EU (since 1996)"]))
```

```python
#  We read into our dataframe the detailed Data for German movies, minority coproductions included
DataGermanMoviesMinorityIncluded = pd.DataFrame([])
for i in range(0, 200): 
    path = working_directory + "/DatengrundlageEUOBS/Deutsche_Filme Daten_Minority_Co-Productions_included/Deutsche_Filme_Details/lumiere-film(" + str(i) + ").csv"
    locals()["dflocal" + str(i)]= pd.read_csv(path, sep=";", engine="python") 
    
    # we now add an index to each local dataframe to know all entries refer to the same movie later
    locals()["dflocal" + str(i)]["index"] = i 
    
    # we now define a variable with the number of columns in each table 
    # and the rename the "total since n" column to "total"
    #we also drop all yearly columns to make it uniform
    locallength = len(locals()["dflocal" + str(i)].columns)
    locals()["dflocal" + str(i)].rename(columns={locals()["dflocal" + str(i)].columns[(locallength-2)]: "Total" }, inplace = True)
    locnum = list(range(3, (locallength-2)))
    locals()["dflocal" + str(i)].drop(locals()["dflocal" + str(i)].columns[locnum], axis=1, inplace=True)
    DataGermanMoviesMinorityIncluded = pd.concat([DataGermanMoviesMinorityIncluded, locals()["dflocal" + str(i)]])
   
```

```python
#  WE GROUP THE MOVIE DETAILS BY INDEX, THEN COMBINE EACH GROUPING WITH THE MOVIE TITLE OF 
# THE CORRESPONDING INDEX
DataGermanMoviesMinorityIncludedGrouped = DataGermanMoviesMinorityIncluded.groupby("index")
for i in range(0, 200): 
    LOCDF = (pd.DataFrame(GermanMoviesMinorityIncluded.iloc[i])).T
    locals()["CombinedGermanMoviesMinorityIncluded" + str(i)] = pd.concat([LOCDF,DataGermanMoviesMinorityIncludedGrouped.get_group(i)], axis=0, ignore_index=True)
```

```python
#   WE NOW CLEAN THE COMBINED DATAFRAMES FOR EACH MOVIE
# WE REPLACE NaN VALUES In "TOTAL" COLUMN with 0, remove empty spaces 
# and covert to int datatype by rounding 
for i in range(0, 200): 
    (locals()["CombinedGermanMoviesMinorityIncluded"+str(i)]['Total']) = (locals()["CombinedGermanMoviesMinorityIncluded"+str(i)]['Total']).fillna(0)
    (locals()["CombinedGermanMoviesMinorityIncluded"+str(i)]['Total']) = (locals()["CombinedGermanMoviesMinorityIncluded"+str(i)]['Total']).str.replace(' ', '').astype("float64")
    (locals()["CombinedGermanMoviesMinorityIncluded"+str(i)]['Total']) = (locals()["CombinedGermanMoviesMinorityIncluded"+str(i)]['Total']).fillna(0)
    (locals()["CombinedGermanMoviesMinorityIncluded"+str(i)]['Total']) = (locals()["CombinedGermanMoviesMinorityIncluded"+str(i)]['Total']).round(0).astype("int")
```

```python
#   We now add up the total admissions in FR for German movies, minority productions included
# Via the "try" and "except" methods, we skip those movies where data/sales in France do not exist. 

totalinFRGermanMoviesMinorityIncluded = 0

for i in range(0, 200):
    try:
        localN = int((locals()["CombinedGermanMoviesMinorityIncluded"+str(i)].loc[locals()["CombinedGermanMoviesMinorityIncluded"+str(i)]['Market'] == "FR"])["Total"])
        totalinFRGermanMoviesMinorityIncluded = totalinFRGermanMoviesMinorityIncluded + localN
    except TypeError as verr:
        pass


print("Total admissions of German movies in France (minority coproductions included) is", totalinFRGermanMoviesMinorityIncluded)
```

```python
#    We now add up the total admissions in DE for German movies, minority productions included
# Via the "try" and "except" methods, we skip those movies where data/sales in Germany do not exist. 

totalinDEGermanMoviesMinorityIncluded = 0

for i in range(0, 200):
    try:
        localN = int((locals()["CombinedGermanMoviesMinorityIncluded"+str(i)].loc[locals()["CombinedGermanMoviesMinorityIncluded"+str(i)]['Market'] == "DE"])["Total"])
        totalinDEGermanMoviesMinorityIncluded = totalinDEGermanMoviesMinorityIncluded + localN
    except TypeError as verr:
        pass


print("Total admissions of German movies in Germany (minority coproductions included) is", totalinDEGermanMoviesMinorityIncluded)
```

```python
#   Percentage of admissions in France or Germany for German movies, minority coproductions included

PercentDEGermanyAdmissionsMinorityIncluded = totalinDEGermanMoviesMinorityIncluded/SUMGermanMoviesMinorityIncluded
print("German movies, minority productions included make", PercentDEGermanyAdmissionsMinorityIncluded, "% of their total EU admissions in Germany")

PercentFRGermanyAdmissionsMinorityIncluded = totalinFRGermanMoviesMinorityIncluded/SUMGermanMoviesMinorityIncluded
print("German movies, minority productions included make", PercentFRGermanyAdmissionsMinorityIncluded, "% of their total EU admissions in France")


```

```python
#   Number of films actually running in other country

AnzahlinFRGermanMoviesMinorityIncluded = 0

for i in range(0, 200):
    try:
        int((locals()["CombinedGermanMoviesMinorityIncluded"+str(i)].loc[locals()["CombinedGermanMoviesMinorityIncluded"+str(i)]['Market'] == "FR"])["Total"])
        AnzahlinFRGermanMoviesMinorityIncluded = AnzahlinFRGermanMoviesMinorityIncluded + 1
    except TypeError as verr:
        pass
    
print("Number of German movies, (minority coproductions included), that actually ran in France", AnzahlinFRGermanMoviesMinorityIncluded)
print()
durchschnittinFRGermanMoviesMinorityIncluded = (totalinFRGermanMoviesMinorityIncluded / AnzahlinFRGermanMoviesMinorityIncluded)
print("Their average admissions in France (per film that's actually running there) is:", durchschnittinFRGermanMoviesMinorityIncluded)
```

```python
#   constructing contracted dataframe with total earnings in both countries so that we may plot it
# for this, we add the columns "Totale in DE", "Total in FR" and years 1-20 for both admissions in Germany and France for each movie from the
# oversight dataframe created at the beginning. 

GermanMoviesMinorityIncluded["Total in FR"] = "Total in FR"
GermanMoviesMinorityIncluded["Total in DE"] = "Total in DE"

for i in range(0, 200): 
    try:
        localN1 = int((locals()["CombinedGermanMoviesMinorityIncluded"+str(i)].loc[locals()["CombinedGermanMoviesMinorityIncluded"+str(i)]['Market'] == "FR"])["Total"])
    except TypeError as verr:
        localN1 = 0 
        pass
    
    
    try:
        localN2 = int((locals()["CombinedGermanMoviesMinorityIncluded"+str(i)].loc[locals()["CombinedGermanMoviesMinorityIncluded"+str(i)]['Market'] == "DE"])["Total"])
    except TypeError as verr:
        localN2 = 0 
        pass
  
    GermanMoviesMinorityIncluded.at[i,"Total in FR"]= localN1
    GermanMoviesMinorityIncluded.at[i,"Total in DE"] = localN2
```
